```
#Data Modelling - Week 3
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```
#Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,
                                                 y,
                                                test_size=0.2,
                                                random_state =5)
```

```
#Check the dimensions
X_train.shape
```

```
#Create model
#As data is numerical, logistic regression could be used
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```
print(model.score(X_train,y_train))
print(model.score(X_test,y_test))
```

```
#Confusion matrix can be used to check  accuracy.
from sklearn.metrics import confusion_matrix
confusion_matrix1 = confusion_matrix(y,model.predict(x))
confusion_matrix1
```

```
from sklearn.metrics import classification_report
print(classification_report(y,model.predict(x)))
```

```
#Preparing ROC Curve (Receiver Operating Characteristics Curve)
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

# predict probabilities
probs = model.predict_proba(x)
# Consider positive outcome only
probs = probs[:, 1]
# Calculate AUC
auc = roc_auc_score(y, probs)
print('AUC: %.3f' % auc)
# calculate roc curve
fpr, tpr, thresholds = roc_curve(y, probs)
plt.plot([0, 1], [0, 1], linestyle='--')
# Plot the roc curve 
plt.plot(fpr, tpr, marker='.')
```

```
#Decision tree
#Applying Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
model_desc = DecisionTreeClassifier(max_depth=5)
model_desc.fit(X_train,y_train)
```

```
model_desc.score(X_train,y_train)
```

```
model_desc.score(X_test,y_test)
```

```
#Applying Random Forest
from sklearn.ensemble import RandomForestClassifier
model_random = RandomForestClassifier(n_estimators=11)
model_random.fit(X_train,y_train)
```

```
model_random.score(X_train,y_train)
```

```
model_random.score(X_test,y_test)
```

```
#Random Forest gave better results
```

```
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier(n_neighbors=7,
                             metric='minkowski',
                             p = 2)
model_knn.fit(X_train,y_train)
```

```
model_knn.score(X_test,y_test)
```

```
#Preparing ROC curve
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

# predict probabilities
probs = model_knn.predict_proba(x)
# keep probabilities for the positive outcome only
probs = probs[:, 1]
# calculate AUC
auc = roc_auc_score(y, probs)
print('AUC: %.3f' % auc)
# calculate roc curve
fpr, tpr, thresholds = roc_curve(y, probs)
print("True Positive Rate - {}, False Positive Rate - {} Thresholds - {}".format(tpr,fpr,thresholds))
# plot no skill
plt.plot([0, 1], [0, 1], linestyle='--')
# plot the roc curve for the model
plt.plot(fpr, tpr, marker='.')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
```

```
#Recall Curve for Logistic Regression
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import f1_score
from sklearn.metrics import auc
from sklearn.metrics import average_precision_score
# predict probabilities
probs = model.predict_proba(x)
# keep probabilities for the positive outcome only
probs = probs[:, 1]
# predict class values
yhat = model.predict(x)
# calculate precision-recall curve
precision, recall, thresholds = precision_recall_curve(y, probs)
# calculate F1 score
f1 = f1_score(y, yhat)
# calculate precision-recall AUC
auc = auc(recall, precision)
# calculate average precision score
ap = average_precision_score(y, probs)
print('f1=%.3f auc=%.3f ap=%.3f' % (f1, auc, ap))
# plot no skill
plt.plot([0, 1], [0.5, 0.5], linestyle='--')
# plot the precision-recall curve for the model
plt.plot(recall, precision, marker='.')
```
